### Problem 4
#### Write a function that returns the count of positive integers k such that 1 <= k < n and k > (n - 1) / 2. (Example: n=9 -> 4).

```python
def count_k(n):
    count = 0
    for k in range(1, n):
        if k > (n - 1) / 2:
            count += 1
    return count
n = int(input("Nhập số n: "))
result = count_k(n)
print("Số lượng số k thỏa mãn là:", result)
```

```python
a, b, c = map(int, input("Nhập một số bất kỳ: "). split())
def tam_giac(a, b, c):
    if (a > 0 and b > 0 and c > 0) and (a + b > c) and (a + c > b) and (b + c > a):
        return True
    else:
        return False
result= tam_giac(a, b, c)
print(result)
```

### Problem 8
#### Write a function that returns the character that is n positions after character x in the alphabet.
#### Preserve case and wrap around within A..Z or a..z.

```python
def shift_char(x, n):
    if x.islower():
        return chr((ord(x) - ord('a') + n) % 26 + ord('a'))
    elif x.isupper():
        return chr((ord(x) - ord('A') + n) % 26 + ord('A'))
    return x

x = input("Nhập ký tự x: ")
n = int(input("Nhập số vị trí n: "))
print("Ký tự sau khi dịch chuyển là:", shift_char(x, n))
```

### Problem 9
#### Write a function that converts an uppercase letter to lowercase. If the input is not a single character, raise ValueError.

```python
def to_lowercase(char):
    if len(char) != 1:
        raise ValueError("Đầu vào phải là một ký tự duy nhất.")
    return char.lower()
char = input("Nhập một chữ cái in hoa: ")
print("Chữ cái in thường là:", to_lowercase(char))
```

### Problem 12
#### Write a function that returns a right isosceles triangle of height n using '*'.
#### Line i (1-indexed) has i '*', joined with newlines.
#### Example with n = 4
#### *
#### **
#### ***
#### ****

```python
def draw_triangle(n):
    lines = []
    for i in range(1, n + 1):
        lines.append("*" * i)
    return "\n".join(lines)
n = int(input("Nhập chiều cao tam giác n: "))
print("Kết quả in ra tam giác:")
print(draw_triangle(n))  
```
