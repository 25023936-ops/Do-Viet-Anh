W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.

```python
s = input()
def detect_number_type(s):
    try:
        int(s)
        return 'int'
    except ValueError:
        try:
            float(s)
            return 'float'
        except ValueError:
            return 'neither'
print(detect_number_type(s))
```

W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.

```python
def are_collinear(x1,y1,x2,y2,x3,y3):
    b = (y2 - y1) * (x3 - x2) - (y3 - y2) * (x2 - x1)
    if b == 0:
        return True
    else:
        return False
x1,y1,x2,y2,x3,y3=map(int,input().split())  
are_collinear(x1,y1,x2,y2,x3,y3) 
```

W5A25 Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3), (0,3)], p = (2,1)

```python
def  point_in_quadrilateral(quad, p):
    def cross_product(o, a, b):
        return (a[0] - o[0]) * (b[1] - o[1]) - (a[1] - o[1]) * (b[0] - o[0])
    px, py = p
    signs = []
    for i in range(4):
        p1 = quad[i]
        p2 = quad[(i + 1) % 4]
        cp = cross_product(p1, p2, p)
        signs.append(cp)
    is_all_non_negative = all(s >= 0 for s in signs)
    is_all_non_positive = all(s <= 0 for s in signs)
    return is_all_non_negative or is_all_non_positive
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
def goc(a,b,c,d,e,f):
    ba_x, ba_y = a - c, b - d
    bc_x, bc_y = e - c, f - d
    vo_huong= ba_x * bc_x + ba_y * bc_y
    ba = (ba_x**2 + ba_y**2)**0.5
    bc = (bc_x**2 + bc_y**2)**0.5
    if ba == 0 or bc == 0:
        return 0
    cos_abc = vo_huong / (ba * bc)
    goc_abc=(2*(1-cos_abc))**0.5
    return goc_abc
a,b,c,d,e,f=map(int,input().split())
goc(a,b,c,d,e,f)
```

W5A33 Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa,

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    if car_type == 4:
        price1 = 11000
        price2 = 13000
        price3 = 10.5000
    else: 
        price1 = 12000
        price2 = 14000
        price3 = 12000
    if km <= 0.8:
        price4 = price1
    elif km <= 20:
        price4 = price1 + (km - 0.8) * price2
    else:
        price4 = price1 + 19 * price2 + (km - 20) * price3
    price4 = 0
    if wait_minutes > 5:
        price5 = (wait_minutes - 5) * 3000
    all_price = price4 + price5
    if (0 <= start_hour <= 5) or start_hour == 22 or start_hour == 23:
        all_price = all_price * 1.2
    if all_price< 50000 and km>30:
        return "50000 VNĐ"
    return round(all_price)
car_type, start_hour, km, wait_minutes=map(int,input().split())
print(smart_taxi_fare(car_type, start_hour, km, wait_minutes))
```

W5A34 Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau. Nếu n ≤ 0 trả về 0

```python
def sum_1_to_n(n):
    if n<=0:
        return 0
    tong=0
    for i in range(1,n+1):
        a=i
        chuso=[]
        while a>0:
         b=a%10
         a=a//10
         chuso.append(b)
        if len(chuso)==len(set(chuso)):
         tong+=i
    return tong         
n=int(input())
sum_1_to_n(n)
```
