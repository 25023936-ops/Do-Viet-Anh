```python
a = int(input("Nhập a: "))
la_chinh_phuong = False

for i in range(1, a + 1):
    if i * i == a:
        la_chinh_phuong = True
        break
    elif i * i > a:  # Vượt quá rồi thì dừng luôn cho nhanh
        break

if la_chinh_phuong:
    print(a, "là số chính phương")
else:
    print(a, "không phải là số chính phương")
```

```python
a = int(input("Nhập a: "))
tong_uoc = 0

for i in range(1, a):
    if a % i == 0:
        tong_uoc += i

if tong_uoc == a and a > 0:
    print(a, "là số hoàn hảo")
else:
    print(a, "không phải là số hoàn hảo")
```

```python
n = int(input("Nhập n: "))
S = 0
k = 0

while True:
    if S + (k + 1) >= n:
        break
    k += 1
    S += k

print("k cần tìm là:", k)
```

```python
A = float(input("Nhập A: "))
S = 0
n = 0

while S <= A:
    n += 1
    S += 1 / n

print("n nhỏ nhất là:", n)
```

```python
max_val = None
min_val = None

while True:
    x = int(input("Nhập số (nhập -1 để dừng): "))
    if x == -1:
        break
    
    if max_val is None or x > max_val:
        max_val = x
    if min_val is None or x < min_val:
        min_val = x

if max_val is not None:
    print(f"Số lớn nhất: {max_val}, Số nhỏ nhất: {min_val}")
```

```python
n = int(input("Nhập n: "))
dem = 0

if n == 0:
    dem = 1
else:
    while n > 0:
        dem += 1
        n //= 10  # Loại bỏ chữ số cuối cùng

print("Số chữ số của n là:", dem)
```

```python
n = int(input("Nhập n: "))
chan = 0
le = 0

if n == 0:
    chan = 1
else:
    while n > 0:
        chu_so = n % 10  # Lấy chữ số cuối
        if chu_so % 2 == 0:
            chan += 1
        else:
            le += 1
        n //= 10

print(f"Chữ số chẵn: {chan}, Chữ số lẻ: {le}")
```

```python
n = int(input("Nhập n: "))
tong = 0

while n > 0:
    tong += n % 10
    n //= 10

print("Tổng các chữ số là:", tong)
```

```python
n = int(input("Nhập n: "))
la_luy_thua = True

if n <= 0:
    la_luy_thua = False
else:
    while n > 1:
        if n % 3 != 0:
            la_luy_thua = False
            break
        n //= 3

if la_luy_thua:
    print("n là số dạng 3^k")
else:
    print("n không phải là số dạng 3^k")
```

```python
A = int(input("Nhập A: "))
f1 = 1
f2 = 1
f_max = 1

while f2 <= A:
    f_max = f2
    
    f_next = f1 + f2
    f1 = f2
    f2 = f_next

print("Số Fibonacci lớn nhất không vượt quá A là:", f_max)
```
