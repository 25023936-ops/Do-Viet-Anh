W4A18 Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b

```python
a, b = map(int,input().split())
uoc_a= []
uoc_b = []
uoc_chung = []
for i in range (1, a+1):
    if a%i == 0:
        uoc_a.append(i)
for j in range (1, b+1):
    if b%j == 0:
        uoc_b.append(j)
for H in (uoc_a):
    if H in (uoc_b):
        uoc_chung.append(H)
print(uoc_a)
print(uoc_b)
print(uoc_chung)
```

W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

W4A25 Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
max_val = None
min_val = None

while True:
    n = int(input("Nhập số: "))
    if n == -1:
        break
        
    if max_val == None:
        max_val = n
        min_val = n
    else:
        if n > max_val:
            max_val = n
        if n < min_val:
            min_val = n

if max_val == None:
    print("Không có số nào được nhập.")
else:
    print(f"{min_val} là số nhỏ nhất")
    print(f"{max_val} là số lớn nhất")
```

W4A26 [Fibonaci1] Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
A = int(input("Nhập số nguyên dương A: "))
a = 1
b = 1
if A < 1:
    print("Không có số Fibonacci nào phù hợp.")
elif A == 1:
    print("Số Fibonacci lớn nhất không vượt quá A là: 1")
else:
   Fi = a+b
while Fi <= A:
    a = b
    b = Fi
    Fi = a+b
print("Số Fibonacci lớn nhất không vượt quá", A, "là:", b)
```
