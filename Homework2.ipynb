```python
# Bai 5
a, b, c= map(int, input("a va b va c bang: ").split())
def equation(a, b, c):
    if a == 0:
            if b == 0:
                if c == 0:  
                    print("Phương trình vô số nghiệm")
                else:
                    print("Phương trình vô nghiệm")
            else:
                print(f"x={-c/b}") 
    else:
        delta = b**2 - 4*a*c
        if delta < 0:
             print("Vo nghiem")
        elif delta == 0:
             print(f"x = {-b / (2*a)}")
        else:
             print(f"x1 = {(-b + delta**0.5) / (2*a)}")
             print(f"x2 = {(-b - delta**0.5) / (2*a)}")
equation(a, b, c)
```

```python
# Bai 8
a, b, m = map(float, input("Nhập a, b, m: ").split())
c, d, n = map(float, input("Nhập c, d, n: ").split())


D = a * d - b * c
Dx = m * d - b * n
Dy = a * n - m * c

if D == 0:
    if Dx == 0 and Dy == 0:
        print("Hệ phương trình có vô số nghiệm")
    else:
        print("Hệ phương trình vô nghiệm")
else:
    x = Dx / D
    y = Dy / D
    print(f"Hệ có nghiệm duy nhất: x = {x}, y = {y}")
```

```python
# Bai 10


x_c, y_c = map(float, input("Nhập tọa độ tâm (x_c, y_c): ").split())
r = float(input("Nhập bán kính r: "))


x, y = map(float, input("Nhập tọa độ điểm A(x, y): ").split())


d_squared = (x - x_c)**2 + (y - y_c)**2
r_squared = r**2

if d_squared <= r_squared:
    print("Điểm A thuộc đường tròn (nằm trong hoặc trên đường viền).")
else:
    print("Điểm A không thuộc đường tròn (nằm ngoài).")
```
