Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2. Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này. Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn.

```python
def hamming(x: int, y: int) -> int:
    xor_result = x ^ y
    return bin(xor_result).count('1')
x,y=map(int,input().split())
print(f"Khoảng cách Hamming giữa {x} và {y} là: {hamming(x, y)}")
```

Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành 'z', 'b' thành 'b' và 'c' thành 'x'. Viết hàm trả về (true) nếu hai từ (a) và (b) là đẳng cấu, ngược lại (false). Bạn có thể giả sử hai từ đầu vào có cùng độ dài.

```python
def dangcau(a, b):
    chuanhxa = {}
    ketquaanhxa = set()
    for i,j in zip(a, b):
        if i in chuanhxa:
            if chuanhxa[i] !=j:
                return "không phải đẳng cấu"
        else:
            if j in ketquaanhxa:
                return "không phải đẳng cấu"
            chuanhxa[i] = j
            ketquaanhxa.add(j)
    return "là đẳng cấu"
a,b=map( str,input().split())
print(dangcau(a,b))
```

Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và 0 <= |r| < |divisor|.

```python
def divide_with_remainder(dividend, divisor):
    if divisor == 0:
        return "nhập lại"
    q = dividend // divisor
    r = dividend % divisor
    return q, r
a,b=map(int,input().split())
print(*divide_with_remainder(a, b))
```

Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông

```python
def is_right_triangle(a, b, c):
    if a**2+b**2==c**2 or a**2 +c**2==b**2 or a**2==b**2+c**2:
        return True
    else:
        return False
a,b,c=map(int,input().split())
is_right_triangle(a,b,c)
```
